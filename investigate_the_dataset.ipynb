```python
from voc12 import dataloader
from voc12.dataloader import TorchvisionNormalize
from PIL import Image
import numpy as np
from skimage.color import label2rgb
```

```python
voc12_root = '../data/raw/VOCdevkit/VOC2012'
train_list = 'voc12/val.txt'
val_list = 'voc12/val.txt'
cls_label_path = 'voc12/cls_labels.npy'
label_dir= '../data/preprocessed/cv/slic/'
crop_size= 512
```

```python
dataset_train = dataloader.VOC12SegmentationDataset(
        train_list, label_dir, crop_size, voc12_root,
        rescale=None, img_normal=None, hor_flip=False,
        crop_method='random', resize_long=(320, 640)
    )
```

```python
counter = 0
for i in dataset_train:
    img = i['img']
    seg = i['seg_label']
    display(Image.fromarray(img.transpose(1, 2, 0).astype(np.uint8()), 'RGB'))
    display(Image.fromarray((label2rgb(seg, bg_label=0) * 255).astype(np.uint8()), 'RGB'))
    if counter == 6:
        break
    counter += 1
```
